```python
import pandas as pd
```

```python
import glob
```

```python
df_list = []
files_list = glob.glob('*.csv')
for file in files_list:
    df = pd.read_csv(file)
    df_list.append(df)
```

```python
df=pd.concat(df_list)
```

```python
df.shape
```

```python
df
```

```python
df["regname"].value_counts()
```

```python
df=df[df["regname"]=="Mains"]
```

```python
df.shape
```

```python
df.head()
```

```python
df=df.drop(columns=["max_t","n","apparent_power","pf","regname"])
```

```python
df.head()
```

```python
df.rename(columns={"ee_site_id":"site_id","min_t":"Timestamp"},inplace=True)
```

```python
df.head()
```

```python
df['Timestamp'] = pd.to_datetime(df.Timestamp)
```

```python

df_pivoted = df.pivot(index='Timestamp', columns='site_id', values='power')

df_pivoted = df_pivoted.reset_index()
df_pivoted
```

```python
df= df_pivoted.rename_axis(None, axis=1)
```

```python
df.index = df.Timestamp
df
```

```python
df.drop(columns='Timestamp', axis=1, inplace=True)
df
```

```python
df.index=pd.to_datetime(df.index)
df_hour=df.resample("h").sum()
df_hour
```
