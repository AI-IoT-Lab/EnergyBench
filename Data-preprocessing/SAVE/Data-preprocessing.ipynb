```python
files_list = glob.glob('save_consumption_data_2018_2_v0-1/save_consumption_data_2018_2_v0-1/*.csv')
files_list
```

```python
for file in files_list:
    file_name = os.path.basename(file).split('.csv')[0]
    df = pd.read_csv(file)
    df.recorded_timestamp = pd.to_datetime(df['recorded_timestamp'], unit='s')
    df = df.sort_values("recorded_timestamp")
    print(file, f"No. of buildings: {df.bmg_id.unique().shape}")
    df.bmg_id = df.bmg_id.apply(lambda x: x[2:])
    final_df = df.pivot(index='recorded_timestamp', columns='bmg_id', values='energy')
    final_df = final_df.reset_index()
    final_df = final_df.rename_axis(None, axis=1)
    final_df.rename(columns={'recorded_timestamp': 'Timestamp'}, inplace=True)
    final_df.set_index('Timestamp', inplace=True)
    final_df_new = final_df / 1000
    final_df_new.to_csv(f"/home/user/New_Buildings_Datasets/New-Bs/SAVE/processed/15min/{file_name}-15min.csv")
    final_df_new_1h = final_df_new.resample('1H').sum(min_count=1)
    final_df_new_1h.to_csv(f"/home/user/New_Buildings_Datasets/New-Bs/SAVE/processed/hourly/{file_name}-1H.csv")
```

```python
df = pd.read_csv(f"/home/user/New_Buildings_Datasets/New-Bs/SAVE/processed/hourly/SAVE-1H.csv")
df.set_index("Timestamp", inplace=True)
df.index = pd.to_datetime(df.index)
df
```

```python
df.info()
```

```python
gh = {}
for id in df.columns:
    gh[id] = (df[id].isna().sum() / df.shape[0]) * 100
gh
```

```python
missing_df = pd.DataFrame(gh.items(), columns=['Buildings', 'Percentage'])
missing_df
```

```python
missing_df['Missing values'] = missing_df['Buildings'].apply(lambda x: df[x].isna().sum())
missing_df
```

```python
missing_df['Total values'] = [len(df)] * len(missing_df)
missing_df
```

```python
missing_df.to_csv('missing_values.csv', index=False)
```
