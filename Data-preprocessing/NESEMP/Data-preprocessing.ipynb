```python
import pandas as pd
import os
import glob
import regex as rx
```

```python
file = 'raw_data/feeds-1.csv'
```

```python
files_list = glob.glob('raw_data/*.csv')
files_list
```

```python
df_list = []
for each in files_list:
    filename = os.path.basename(each)
    house = rx.findall('[0-9]+', filename)[0]
    df = pd.read_csv(each)
    df.drop(columns=[df.columns[-1]], inplace=True)
    row = df.columns.to_list()
    row[-1] = int(row[-1])
    df.columns = ['Timestamp', int(house)]
    df.loc[len(df)] = row
    df.Timestamp = pd.to_datetime(df.Timestamp).dt.tz_localize(None)
    df = df.sort_values('Timestamp')
    df.set_index('Timestamp', inplace = True)
    df = df / 1000
    df_list.append(df)
```

```python
df_list
```

```python
len(df_list)
```

```python
df = pd.concat(df_list)
df
```

```python
col_list = df.columns.tolist()
col_list = sorted(col_list)
col_list
```

```python
df_new = df[col_list]
df_new
```

```python
df_new.columns = df_new.columns.astype(str)
df_new
```

```python
df_new_15min = df_new.resample("15min").sum(min_count=1)
df_new_15min
```

```python
df_new_15min.to_csv('/home/user/New_Buildings_Datasets/New-Bs/NESEMP/processed/15min/NESEMP-15min.csv')
```

```python
df_new_1hr = df_new.resample("1H").sum(min_count=1)
df_new_1hr
```

```python
df_new_1hr.to_csv('/home/user/New_Buildings_Datasets/New-Bs/NESEMP/processed/hourly/NESEMP-1H.csv')
```

```python
df = pd.read_csv('/home/user/New_Buildings_Datasets/New-Bs/NESEMP/processed/hourly/NESEMP-1H.csv')
df.set_index("Timestamp", inplace=True)
df
```

```python
gh = {}
for id in df.columns:
    gh[id] = (df[id].isna().sum() / df.shape[0]) * 100
gh
```

```python
missing_df = pd.DataFrame(gh.items(), columns=['Buildings', 'Percentage'])
missing_df
```

```python
missing_df['Missing values'] = missing_df['Buildings'].apply(lambda x: df[x].isna().sum())
missing_df
```

```python
missing_df['Total values'] = [len(df)] * len(missing_df)
missing_df
```

```python
missing_df.to_csv('missing_values.csv', index=False)
```
