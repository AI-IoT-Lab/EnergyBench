```python
import pandas as pd
import os
import glob
```

```python
file = 'csv/electricity_10min.csv'
```

```python
df = pd.read_csv(file)
df
```

```python
meta = 'csv/meta.csv'
```

```python
df_meta = pd.read_csv(meta)
df_meta
```

```python
df['House'] = df.Meta_idMeta.map(df_meta.set_index('idMeta')['Household_idHousehold'])
df
```

```python
df.dt = pd.to_datetime(df.dt)
df.info()
```

```python
final_df = df.pivot_table(index='dt', columns='House', values='Watt')
final_df
```

```python
final_df = final_df.reset_index()
final_df = final_df.rename_axis(None, axis=1)
final_df
```

```python
final_df.rename(columns={'dt': 'Timestamp'}, inplace=True)
final_df
```

```python
final_df.set_index('Timestamp', inplace=True)
```

```python
final_df_new = final_df / 1000
final_df_new
```

```python
final_df_new.columns = final_df_new.columns.astype(str)
final_df_new
```

```python
final_df_30min = final_df_new.resample('30min').sum(min_count=1)
final_df_30min
```

```python
final_df_30min.to_csv("/home/user/New_Buildings_Datasets/New-Bs/METER/processed/30min/METER-30min.csv")
```

```python
final_df_1h = final_df_new.resample('1H').sum(min_count=1)
final_df_1h
```

```python
final_df_1h.to_csv("/home/user/New_Buildings_Datasets/New-Bs/METER/processed/hourly/METER-1H.csv")
```

```python
df = pd.read_csv("/home/user/New_Buildings_Datasets/New-Bs/METER/processed/hourly/METER-1H.csv")
df.set_index("Timestamp", inplace=True)
df
```

```python
gh = {}
for id in df.columns:
    gh[id] = (df[id].isna().sum() / df.shape[0]) * 100
gh
```

```python
missing_df = pd.DataFrame(gh.items(), columns=['Buildings', 'Percentage'])
missing_df
```

```python
missing_df['Missing values'] = missing_df['Buildings'].apply(lambda x: df[x].isna().sum())
missing_df
```

```python
missing_df['Total values'] = [len(df)] * len(missing_df)
missing_df
```

```python
missing_df.to_csv('missing_values.csv', index=False)
```
