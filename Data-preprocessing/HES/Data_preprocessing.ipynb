```python
df_main = df[(df.Appliance == 240) | (df.Appliance == 241)].sort_index()
df_main
```

```python
df_main.reset_index(inplace=True)
```

```python
df_main
```

```python
df_main.Household.unique()
```

```python
for house in df_main.Household.unique():
    df_m = df_main[df_main.Household == house]
    df_m = df_m.pivot(index='Timestamp', columns='Household', values='value')
    df_m = df_m.reset_index()
    df_m = df_m.rename_axis(None, axis=1)
    df_m.set_index('Timestamp', inplace=True)
    df_m_15 = df_m.resample('15min').sum(min_count=1) / 1000
    df_m_15.to_csv(f'/home/user/New_Buildings_Datasets/New-Bs/Household-Electricity-Survey/processed/15min/{house}_15min.csv')
    df_m_1 = df_m.resample('1H').sum(min_count=1) / 1000
    df_m_1.to_csv(f'/home/user/New_Buildings_Datasets/New-Bs/Household-Electricity-Survey/processed/hourly/{house}_1H.csv')
# df_main
```
